```python
def getExtFromURL(url):
    path = urllib.parse.urlparse(url).path
    return os.path.splitext(path)[1]
```

```python
train_dir = data_dir + '/train'
if not os.path.exists(train_dir):
    os.mkdir(train_dir)
```
